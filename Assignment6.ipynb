```python
import numpy as np
import matplotlib.pyplot as plt

# Ask the user to enter a value for M
M = float(input("Enter a value for M: "))

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Compute y values
y1 = M * x**2
y2 = M * np.sin(x)

# Plot both functions in a single figure
plt.plot(x, y1, label='y = M * x^2', color='blue', linestyle='-')
plt.plot(x, y2, label='y = M * sin(x)', color='red', linestyle='--')

# Add legend, grid, and title
plt.legend()
plt.grid(True)
plt.title('Mathematical Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 90, 78, 88, 92]
}

# Convert it into a Pandas DataFrame
df = pd.DataFrame(data)

# Plot the scores using a Seaborn bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')

# Add annotations on top of each bar
for index, row in df.iterrows():
    bar_plot.text(row.name, row.Score + 1, round(row.Score, 2), color='black', ha="center")

# Add title, axis labels, and grid
plt.title('Scores of Different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace with your roll number
roll_number = 123456

# Set seed for NumPy's random number generator
np.random.seed(roll_number)

# Generate a dataset of 50 values using np.random.randn()
data = np.random.randn(50)

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line plot showing cumulative sum
axs[0, 0].plot(np.cumsum(data))
axs[0, 0].set_title('Cumulative Sum')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Sum')
axs[0, 0].grid(True)

# Scatter plot with random noise
noise = np.random.randn(50)
axs[0, 1].scatter(data, noise)
axs[0, 1].set_title('Scatter Plot with Random Noise')
axs[0, 1].set_xlabel('Data')
axs[0, 1].set_ylabel('Noise')
axs[0, 1].grid(True)

# Hide the empty subplots
axs[1, 0].axis('off')
axs[1, 1].axis('off')

# Display the plots
plt.tight_layout()
plt.show()
```
